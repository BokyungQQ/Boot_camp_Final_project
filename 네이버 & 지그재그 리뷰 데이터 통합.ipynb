```python
def add_info(df, product_name):
    df['ìƒí’ˆëª…'] = product_name
    df['ìƒí’ˆëª…_ì˜µì…˜'] = product_name + '_' + df['ìƒ‰ìƒ']
    df['í”Œë«í¼'] = 'ì™¸ë¶€'

add_info(GV, 'GVbag')
GV
```

```python
# í•œë‹¬ì‚¬ìš©ê¸° -> í›„ê¸° ìœ„ì¹˜ ì´ë™
def move_review(df):
    for i in range(len(df)):
        if df['í›„ê¸°'][i] == 'í•œë‹¬ì‚¬ìš©ê¸°':
            df['í›„ê¸°'][i] = df['í•œë‹¬ì‚¬ìš©ê¸°'][i]

move_review(GV)
GV
```

```python
# í‰ê°€ í•­ëª© ì»¬ëŸ¼ ìƒì„±
def new_column(df):
    df['ì‚¬ì´ì¦ˆ'] = ""
    df['ë¬´ê²Œê°'] = ""
    df['ë””ìì¸'] = ""
    df['í€„ë¦¬í‹°'] = ""
    df['ë§ˆê°ì²˜ë¦¬'] = ""

new_column(GV)
GV
```

```python
def process_data(df):
    for i in range(len(df)):
        for j in range(1, 4):  # Assuming there are three sets of columns: í•­ëª©1, í‰ê°€1, í•­ëª©2, í‰ê°€2, í•­ëª©3, í‰ê°€3
            item_col = f'í•­ëª©{j}'
            evaluation_col = f'í‰ê°€{j}'

            if df[item_col][i] == 'ì‚¬ì´ì¦ˆ':
                df['ì‚¬ì´ì¦ˆ'][i] = df[evaluation_col][i]
            elif df[item_col][i] in ["ë¬´ê²Œ", "ë¬´ê²Œê°"]:
                df['ë¬´ê²Œê°'][i] = df[evaluation_col][i]
            elif df[item_col][i] == 'ë””ìì¸':
                df['ë””ìì¸'][i] = df[evaluation_col][i]
            elif df[item_col][i] == 'í€„ë¦¬í‹°':
                df['í€„ë¦¬í‹°'][i] = df[evaluation_col][i]
            elif df[item_col][i] == 'ë§ˆê°ì²˜ë¦¬':
                df['ë§ˆê°ì²˜ë¦¬'][i] = df[evaluation_col][i]
```
