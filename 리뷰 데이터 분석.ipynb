```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from konlpy.tag import Okt # 형태소 분석기 Okt
from collections import Counter
from wordcloud import WordCloud
plt.rc('font', family='NanumBarunGothic')
```

```python
review = pd.read_csv('/content/Review_Final_real.csv')
review['상품명'] = review['상품명'].str.capitalize()
review
```

```python
# 명사만 추출
tokenizer_noun = Okt()
review['tokenized noun'] = review['후기'].apply(tokenizer_noun.nouns)
```

```python
# 형태소만 추출
tokenizer = Okt()
review['tokenized phrases'] = review['후기'].apply(tokenizer.morphs)
review['tokenized phrases']
```

```python
# 어구 추출
tokenizer_phrases = Okt()
review['tokenized phrases'] = review['후기'].apply(tokenizer_phrases.phrases)
```

```python
stop_words = ['네이버','구매','등록','페이','2023-05','2023-06','2023-07','2023-08','2023-09','2023-10',
              '가방','.','...','..','배송','등록된 네이버','등록된 네이버 페이','등록된 네이버 페이 구매평','진짜','아주','정도','이것','저것','가성','완전','정말']
```

# 베스트 셀러 개별 분석

## 1. G bag



```python
g_review = review[review['상품명'] == 'G bag']

g_words = np.hstack(g_review['tokenized phrases'].values)

# 단어 길이가 1보다 큰 것만 필터링
g_filtered_words = [word for word in g_words if len(word) > 1]

# 단어의 빈도수를 계산
g_word_count = Counter(g_filtered_words)
g_word_count = Counter({x : g_word_count[x] for x in g_word_count if len(x) > 1})
g_word_count = Counter({x : g_word_count[x] for x in g_word_count \
                               if x not in stop_words})
g_word_count = sorted(g_word_count.items(), key=lambda x: x[1], reverse=True)

# 정렬된 결과 출력
for word, count in g_word_count:
    print(f"{word}: {count}번")
```

```python
from collections import Counter

# all_word_count 정렬
all_word_count = sorted(all_word_count, key=lambda x: len(x[0]), reverse=True)

# 가장 긴 단어만 남기고 정리
filtered_word_count = {}
for word, count in all_word_count:
    if word not in filtered_word_count:
        is_overlap = False
        # 이미 추가된 단어들과 비교
        for existing_word in filtered_word_count:
            if word in existing_word:
                is_overlap = True
                break
        if not is_overlap:
            filtered_word_count[word] = count

# count가 높은 순으로 정렬
sorted_filtered_word_count = sorted(filtered_word_count.items(), key=lambda x: x[1], reverse=True)

# 정렬된 결과 출력
for word, count in sorted_filtered_word_count:
    print(f"{word}: {count}번")
```
